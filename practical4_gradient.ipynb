```python
x = 2
lr = 0.0047
max_itr = 10000
itr = 0
precision = 0.000001
prev_step_size = 1
gf = lambda x:(x+3)**2
```

```python
gd = []
while precision < prev_step_size and itr < max_itr:
    prev = x
    x = x-lr*gf(prev)
    prev_step_size = abs(x-prev)
    itr += 1
    print("Iterations: ", itr, "Values: ", x)
    gd.append(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(gd)
```
